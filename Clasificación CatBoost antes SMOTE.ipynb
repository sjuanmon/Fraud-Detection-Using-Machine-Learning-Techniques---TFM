```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix
from catboost import CatBoostClassifier
import numpy as np
import pandas as pd

# Cargar datos
df = pd.read_csv("Base.csv")
df.describe()
```

```python
# Convertir a categÃ³ricas
df['payment_type'] = df.payment_type.astype('category')
df['employment_status'] = df.employment_status.astype('category')
df['housing_status'] = df.housing_status.astype('category')
df['source'] = df.source.astype('category')
df['device_os'] = df.device_os.astype('category')
```

```python
# One-hot encoding
df = pd.get_dummies(df, drop_first=True)
```

```python
# Separar X e y
X = df.drop('fraud_bool', axis=1)
y = df['fraud_bool']
```

```python
# Split train/test
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)
```

```python
# Entrenar CatBoost sin SMOTE
cat_model = CatBoostClassifier(verbose=0, random_state=42)
cat_model.fit(X_train, y_train)
```

```python
# Predicciones
y_pred = cat_model.predict(X_test)
```

```python
# EvaluaciÃ³n
print("=== CATBOOST SIN SMOTE ===")
print("Matriz de ConfusiÃ³n:")
print(confusion_matrix(y_test, y_pred))
print("\nReporte de ClasificaciÃ³n:")
print(classification_report(y_test, y_pred))
```

```python
Observaciones:

El modelo predice casi perfectamente la clase 0 (no fraude).

Pero la clase 1 (fraude) es donde mÃ¡s falla: aunque tiene un recall muy bajo (0.03), la precisiÃ³n es algo mÃ¡s decente (0.51).

Esto pasa por el desequilibrio de clases: sin SMOTE ni tÃ©cnicas de balanceo, el modelo aprende a predecir mayoritariamente la clase mayoritaria (0).
```
