```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split

df = pd.read_csv("Base.csv")
df.head()
```

```python
# Convertir variables categóricas a tipo 'category'
df['payment_type'] = df.payment_type.astype('category')
df['employment_status'] = df.employment_status.astype('category')
df['housing_status'] = df.housing_status.astype('category')
df['source'] = df.source.astype('category')
df['device_os'] = df.device_os.astype('category')
```

```python
# Aplicar One-Hot Encoding
df = pd.get_dummies(df, drop_first=True)
```

```python
# ---------------------------------------------------------------------------------------------------------
```

```python
corr_target = df.corr()['fraud_bool'].sort_values(ascending=False)
print(corr_target)
```

```python
🔹 Interpretación:

Si una variable tiene una correlación cercana a 0, no tiene relación con el fraude y puede ser irrelevante.

Variables con correlación entre 0.1 y 0.3 tienen una relación débil, pero pueden ser útiles.

Valores mayores a 0.3 o menores a -0.3 suelen ser más relevantes.
```

```python
# Eliminar las columnas menos relevantes
cols_to_drop = [
    'device_os_x11', 'payment_type_AD', 'employment_status_CG', 
    'source_TELEAPP', 'zip_count_4w', 'session_length_in_minutes', 
    'device_fraud_count', 'month'  # Contiene NaN
]

df.drop(columns=cols_to_drop, inplace=True)
```

```python
# Separar variables predictoras (X) y variable objetivo (y)
X = df.drop(['fraud_bool'], axis=1)
y = df['fraud_bool']
```

```python
# Dividir en conjunto de entrenamiento y conjunto de prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)
```

```python
# Verificar las formas de los conjuntos
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
# Entrenar el modelo con RandomForestClassifier
from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier(random_state=42)
classifier.fit(X_train, y_train)
```

```python
# Realizar predicciones
predictions = classifier.predict(X_test)
```

```python
# Evaluar el modelo: Matriz de confusión y reporte de clasificación
from sklearn.metrics import confusion_matrix, classification_report
```

```python
cm = confusion_matrix(y_test, predictions)
print("Matriz de Confusión:")
print(cm)
```

```python
report = classification_report(y_test, predictions)
print("Reporte de Clasificación:")
print(report)
```

```python
Comparación de Resultados:
Sin Eliminar Variables:
Matriz de Confusión:

Clase 0 (No fraude): 296690 correctos, 1 incorrecto

Clase 1 (Fraude): 3303 incorrectos, 6 correctos

Reporte de Clasificación:

Precisión para la clase 1 (Fraude): 0.86, pero recall es 0.00 (indica que no se está prediciendo correctamente casi nada de la clase 1).

El modelo predice muy bien la clase 0, pero no predice bien la clase 1 (fraude).

Macro promedio de recall es 0.50, lo que indica que el modelo no está equilibrado entre las clases.

Después de Eliminar Variables:
Matriz de Confusión:

Clase 0 (No fraude): 296688 correctos, 3 incorrectos

Clase 1 (Fraude): 3305 incorrectos, 4 correctos

Reporte de Clasificación:

Precisión para la clase 1 (Fraude): 0.57 (mejor que antes), pero recall sigue siendo 0.00 (todavía no predice correctamente los fraudes).

El modelo sigue siendo muy bueno para la clase 0 (no fraude), pero no ha mejorado su capacidad para predecir la clase 1.

El macro promedio de recall sigue siendo bajo (0.50).

Segundo dropeo

Se ha repetido el mismo proceso añadiendo dropeando la variable month, el resultado sigue siendo negativo o incluso peor que sin dropearlo, siendo una
una variable que tiene mucha correlación con 4w_velocity.

```

```python
Análisis de la Comparación:
Precisión de la clase 1 (fraude): Sin eliminar las variables, la precisión es más alta (0.86), pero recall es 0.00, lo que significa que no se están detectando los fraudes. Después de eliminar las variables, la precisión para la clase 1 baja a 0.57, pero el recall aún sigue siendo 0.00. Esto sugiere que, aunque las variables eliminadas podrían haber afectado la precisión, el problema principal sigue siendo la incapacidad del modelo para detectar fraudes.

Impacto de eliminar las variables: Parece que las variables eliminadas no contribuyeron significativamente a mejorar la capacidad del modelo para predecir fraudes, ya que el recall sigue siendo bajo. El modelo podría estar sesgado hacia predecir la clase mayoritaria (no fraude), lo cual es común en problemas de desequilibrio de clases.
```
