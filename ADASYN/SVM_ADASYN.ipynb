```python
import pandas as pd
import numpy as np
from sklearn.svm import LinearSVC
from imblearn.over_sampling import ADASYN
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix

df = pd.read_csv("Base.csv")
df.describe()
```

```python
# Convertir a categóricas
df['payment_type'] = df.payment_type.astype('category')
df['employment_status'] = df.employment_status.astype('category')
df['housing_status'] = df.housing_status.astype('category')
df['source'] = df.source.astype('category')
df['device_os'] = df.device_os.astype('category')
```

```python
# One-hot encoding
df = pd.get_dummies(df, drop_first=True)
```

```python
# Separar X e y
X = df.drop('fraud_bool', axis=1)
y = df['fraud_bool']
```

```python
# Split train/test
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)
```

```python
# Aplicar ADASYN
adasyn = ADASYN(random_state=42)
X_train_adasyn, y_train_adasyn = adasyn.fit_resample(X_train, y_train)
```

```python
# Escalar
scaler = StandardScaler()
X_train_adasyn_scaled = scaler.fit_transform(X_train_adasyn)
X_test_scaled = scaler.transform(X_test)
```

```python
#Entrenamiento
svm_model = LinearSVC(max_iter=10000, random_state=42)
svm_model.fit(X_train_adasyn_scaled, y_train_adasyn)
```

```python
#Predicción
y_pred_svm = svm_model.predict(X_test_scaled)
```

```python
print("=== SVM LINEAL ===")
print("Matriz de Confusión:")
print(confusion_matrix(y_test, y_pred_svm))
print("\nReporte de Clasificación:")
print(classification_report(y_test, y_pred_svm))
```

```python
🔍 Interpretación rápida:
🎯 Recall en fraudes (39%) está entre los más altos que hemos logrado hasta ahora.

⚠️ La precisión en clase 1 sigue muy baja (6%), lo que significa que muchos positivos predichos son falsos.

📉 La clase mayoritaria sufre un poco: 21.360 falsos positivos.

📈 En problemas donde detectar fraude es más importante que equivocarse con algunos no-fraudes, este resultado puede ser aceptable.
```
