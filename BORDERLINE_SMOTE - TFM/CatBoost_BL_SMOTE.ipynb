```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report, confusion_matrix
from imblearn.over_sampling import BorderlineSMOTE
from catboost import CatBoostClassifier

df = pd.read_csv("Base.csv")
df.head()
```

```python
df.describe()
```

```python
# Convertir variables categóricas a tipo 'category'
df['payment_type'] = df.payment_type.astype('category')
df['employment_status'] = df.employment_status.astype('category')
df['housing_status'] = df.housing_status.astype('category')
df['source'] = df.source.astype('category')
df['device_os'] = df.device_os.astype('category')
```

```python
# One-Hot Encoding
df = pd.get_dummies(df, drop_first=True)
```

```python
# Separar variables
X = df.drop('fraud_bool', axis=1)
y = df['fraud_bool']
```

```python
# División del dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, test_size=0.3, random_state=42)
```

```python
# Aplicar Borderline-SMOTE
border_smote = BorderlineSMOTE(random_state=42)
X_train_border, y_train_border = border_smote.fit_resample(X_train, y_train)
```

```python
#Escalado
scaler = StandardScaler()
X_train_border_scaled = scaler.fit_transform(X_train_border)
X_test_scaled = scaler.transform(X_test)
```

```python
# Entrenar modelo CatBoost
catboost_model_border = CatBoostClassifier(verbose =0, random_state=42 )
catboost_model_border.fit(X_train_border_scaled, y_train_border)
```

```python
# Predicciones
y_pred_border = catboost_model_border.predict(X_test_scaled)
```

```python
# Resultados
print("\n=== BORDERLINE-SMOTE ===")
print("Matriz de Confusión:")
print(confusion_matrix(y_test, y_pred_border))
print("\nReporte de Clasificación:")
print(classification_report(y_test, y_pred_border))
```

```python
 Interpretación:
Leve mejora con BorderlineSMOTE en todas las métricas de la clase minoritaria (fraude).

La clase mayoritaria se mantiene fuerte, sin pérdidas importantes en precisión o recall.

Sin embargo, el recall sigue siendo bajo → el modelo aún detecta muy pocos fraudes reales.
```
